```python
# Cleaning the missing data
# listing the null values columns having more than 50%
missing_values=inp_app.isnull().sum()
missing_values=missing_values[missing_values.values>(0.5*len(missing_values))]
len(missing_values)
```

```python
inp_app= inp_app.loc[:, inp_app.isnull().mean() <= 0.5]
inp_app.shape
```

```python
inp_app.isnull().sum()*100/len(inp_app)
```

**Imputation in Categorical Variable columns where missing values are less than around 13%**

```python
#Missing values in Categorical variables NAME_TYPE_SUITE should be replaced with the MODE value 'Unaccompanied'
mode=inp_app.NAME_TYPE_SUITE.mode()
inp_app['NAME_TYPE_SUITE'].fillna(inp_app.NAME_TYPE_SUITE.mode(), inplace=True)
```

**Imputation in numerical variables**

```python
#AMT_REQ_CREDIT_BUREAU_HOUR,AMT_REQ_CREDIT_BUREAU_DAY,AMT_REQ_CREDIT_BUREAU_WEEK,AMT_REQ_CREDIT_BUREAU_MON
#have more than 90% values as 0 (0 enquiries 1 hour, 1 week and 1 month before application),
#we can impute the missing values in these column as 0 (0 is the median for these columns)  

#Missing values in AMT_GOODS_PRICE could be imputed by mean value for this var since this is a continuous float var
Avg=inp_app.AMT_REQ_CREDIT_BUREAU_YEAR.mean()
inp_app['AMT_GOODS_PRICE'].fillna(Avg, inplace=True)
```

```python
#Imputing Missing values in the following columns with 0. 
#We are assuming that missing means that there were no inquiries for the person.
inp_app[['AMT_REQ_CREDIT_BUREAU_HOUR','AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK','AMT_REQ_CREDIT_BUREAU_MON',
         'AMT_REQ_CREDIT_BUREAU_QRT','AMT_REQ_CREDIT_BUREAU_YEAR']]= inp_app[['AMT_REQ_CREDIT_BUREAU_HOUR',
        'AMT_REQ_CREDIT_BUREAU_DAY','AMT_REQ_CREDIT_BUREAU_WEEK' ,'AMT_REQ_CREDIT_BUREAU_MON','AMT_REQ_CREDIT_BUREAU_QRT',
        'AMT_REQ_CREDIT_BUREAU_YEAR']].fillna(value=0.0)
```

**Checking imbalance**

```python
count1 = 0 
count0 = 0
for i in inp_app['TARGET'].values:
    if i == 1:
        count1 += 1
    else:
        count0 += 1
        
count1 = (count1/len(inp_app['TARGET']))*100
count0 = (count0/len(inp_app['TARGET']))*100

x = ['Defaulted Population(TARGET=1)','Non-Defauted Population(TARGET=0)']
y = [count1, count0]

explode = (0.1, 0)  # only "explode" the 1st slice

fig1, ax1 = plt.subplots()
ax1.pie(y, explode=explode, labels=x, autopct='%1.1f%%',
        shadow=True, startangle=110)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Data imbalance',fontsize=25)
plt.show()
```

```python
def plot_prev_bivarContCat(var1,var2):
    plt.figure(figsize=(15,10))
    ax = sns.boxplot(x=var1, y=var2, data=inp_prev,  showfliers = False)
    #plt.title('Prev Loans applications')
    plt.xlabel(var1, fontsize= 20, color='green')
    plt.ylabel(var2, fontsize= 20, color='green')
    plt.xticks(rotation=90)
    ax.tick_params(axis='x', labelsize=20)
    ax.tick_params(axis='y', labelsize=20)
    plt.show()
```
